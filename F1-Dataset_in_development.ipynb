```python
import requests
import pandas as pd
import time
import json
```

### Data from pitstops

```python
   
lista_pitstops = []
lista_ano = []
lista_conductor = []

anos=[2011,2012,2013,2014,2015,2016,2017,2018,2019,2020,2021,2022]

for ano in range(len(anos)):
    ano_=anos[ano]
    url = "http://ergast.com/api/f1/"+str(ano_)+"/5/pitstops.json"
    response = requests.get(url).json()
    pitstops = response["MRData"]["RaceTable"]["Races"][0]["PitStops"]
    time.sleep(1)
    for index in range(len(pitstops)):
            duracion=pitstops[index]["duration"]
            driver_id = pitstops[index]["driverId"]
            season=  response["MRData"]["RaceTable"]["season"]
            lista_pitstops.append(duracion)
            lista_ano.append(season)
            lista_conductor.append(driver_id)
            

            
lists = [lista_ano, lista_conductor, lista_pitstops]
df_pitstops = pd.concat([pd.Series(x) for x in lists], axis=1)            

   
    
```
